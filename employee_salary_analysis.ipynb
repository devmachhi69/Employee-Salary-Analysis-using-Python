```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("emp.csv")
df.head()
```

```python
avg_salary = df.groupby("Department")["Salary"].mean().sort_values(ascending=False)
print(avg_salary)
```

```python
top_emp = df.loc[df["Salary"].idxmax()]
print(top_emp)
```

```python
plt.figure(figsize=(6, 4))
plt.hist(df["Salary"], bins=5, color="skyblue", edgecolor="black")
plt.title("Salary Distribution of Employees")
plt.xlabel("Salary")
plt.ylabel("Number of Employees")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
avg_salary.plot(kind="bar", color="pink")
plt.title("Average Salary by Department")
plt.xlabel("Department")
plt.ylabel("Average Salary")
plt.tight_layout()
plt.show()
```
